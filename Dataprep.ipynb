```python
## Data preparation for the validation dataset

import os
import sys
import numpy as np 
import pandas as pd
import shutil
```

```python
datapath="./data/tiny-imagenet-200/"
train_datapath = datapath + "train/"
test_datapath = datapath + "test/"
val_datapath = datapath + "val/"
```

```python
(os.listdir(val_datapath))
```

```python
text_file = val_datapath+"val_annotations.txt"
```

```python
val_images = val_datapath +"images"
```

```python
text = open(text_file,"r")
```

```python
data = text.readlines()
```

```python
data
```

```python
target_and_image = {}
for line in data:
    words = line.split('\t')
    target_and_image[words[0]]=words[1]
    
```

```python
target_and_image.items()
```

```python
for image, folder in target_and_image.items():
    new_val_datapath= val_datapath + folder
    #if the new folder path doesnt exit then make one and if it exists 
    #put the images in the folder
    if not os.path.exists(new_val_datapath):
        os.makedirs(new_val_datapath)
        source = val_images+"/"+image
        destination = new_val_datapath +"/"+image
        shutil.move(source,destination)
    else:
        source = val_images+"/"+image
        destination = new_val_datapath +"/"+image
        shutil.move(source,destination)
```
